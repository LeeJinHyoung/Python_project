```python
def f(x):
    y = (x-2)*x*(x+2)
    return y

def f2(x,w):
    y = (x-w)*x*(x+2)
    return y

def f3(x,w,k):
    y = (x-w)*x*(x+k)
    return y


#print('x = 1', f(1))
#print('x = 2', f(2))
#print('x = 3', f(3))

x = np.array([1,2,3]) #list -> np.ndarray
x = np.linspace(-3,3,50) #start, finish, samples 
#print('x[] = {}, type = {}'.format(x,type(x)))
#print(f(x))

plt.figure(figsize=(5,5))
plt.plot(x,f(x),'tomato', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')

plt.plot(x,f2(x,w=0),'skyblue', linewidth = 3, label = '$y=(x-0)*x*(x+2)$')
plt.plot(x,f2(x,w=4),'black', linewidth = 3, label = '$y=(x-4)*x*(x+2)$')

plt.plot(x,f3(x,w=2,k=4),'orange', linewidth = 3, label = '$y=(x-0)*x*(x+4)$')

plt.xlabel('&x&')
plt.ylabel('&y&')


plt.grid(True)
plt.legend(loc='upper left')
plt.ylim(-15,15)
plt.xlim(-3,3)
plt.plot(x,f(x),'red', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,3))
plt.subplots_adjust(wspace=0.5, hspace = 0.5)

for i in range(6):
    plt.subplot(2,3,i+1) #plot 묘사의 위치 지정
    plt.title(i+1)
    plt.plot(x,f2(x,i),'k') ##알아보기
    plt.ylim(-20,20)
    plt.grid(True)
plt.show()
```

```python
def f4(x0,x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

#x0, x1에 따라서 각각 f3 함수를 거쳐서 나오는 값을 계산
xn = 50
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
y = np.zeros((len(x0),len(x1)))
#print('x0 = {}'.format(x0))
#print('x1 = {}'.format(x1))
#print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f4(x0[i0],x1[i1])
        
y =np.round(y,1)
#print('y = {}'.format(y))
plt.figure(figsize = (3.5,3))
#plt.gray()
plt.bone()

plt.pcolor(y) #heat map 그리기

plt.colorbar()
plt.show()

plt.savefig('gauss.png')
```
